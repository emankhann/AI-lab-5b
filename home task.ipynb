```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Dataset
data = {
    'Appliances': [1, 2, 3, 4, 5],
    'Energy Consumption': [1, 2, 3, 4, 5]
}
df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Scatter plot
plt.xlabel('Appliances')
plt.ylabel('Energy Consumption')
plt.scatter(df['Appliances'], df['Energy Consumption'], color='red', marker='+', label='Data Points')

# Prepare data for linear regression
X = df[['Appliances']]
y = df['Energy Consumption']
reg = LinearRegression()
reg.fit(X, y)

# Coefficients and intercept
print(f"Coefficient: {reg.coef_[0]}")
print(f"Intercept: {reg.intercept_}")

# Plot regression line
y_pred = reg.predict(X)
plt.plot(df['Appliances'], y_pred, color='blue', label='Regression Line')
plt.legend()
plt.show()

# Predict for new data
new_data = pd.DataFrame({'Appliances': [6, 7, 8]})
predicted_energy = reg.predict(new_data)
for appliance, prediction in zip(new_data['Appliances'], predicted_energy):
    print(f"Predicted 'Energy Consumption' value for Appliance {appliance} is {prediction:.2f}")
```
